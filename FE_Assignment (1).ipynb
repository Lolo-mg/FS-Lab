# Importing the libraries

```python
import pandas as pd 
import numpy as np
import category_encoders 
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from category_encoders import BinaryEncoder
```

# Creating simple Data frame :]

```python
data = {
  "id": [10, 20, 30, 40, 50],
  "age": [29, 25, 23, 24,24],
    "name": ["sara","noura","lamia","fai","lama"],
    "salary": [ 1000,1200,3000,5000,4000],
    "his_work":["good","amazing","very good","bad","very bad"]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)
df
```

# Q1: Apply Ordinal encoding.

```python
## from the dictionary I found it easier to make conection to the DataFrame, chossing the name column and give the values
## and this libarary category_encoders is so amazing to make the data fit to the dataframe
import category_encoders 

mapping = [
    {
        'col': 'his_work',
        'mapping':{
            'amazing': 0, 
            'very good': 1, 
            'good': 2,
            'bad':3,
            'very bad':4
        }
    }
]

ordinal_encoder = category_encoders.OrdinalEncoder(
    cols = ['his_work'],
    return_df = True,
    mapping = mapping
)
df['his_work'] = ordinal_encoder.fit_transform(df['his_work'])
df
```

# Q2: Apply One hot encoding.

```python
### what make it the one hot different from ordinal that there isn`t order for the data :)
### One-hot encoding is used to convert categorical variables into a
### format that can be readily used by machine learning algorithms.


df = pd.DataFrame({'class': ['A', 'A', 'B', 'B', 'B', 'B', 'C', 'C'],
                   'score': [25, 12, 15, 14, 19, 23, 25, 29]})
df
```

```python
### ‘ignore’ : When an unknown category is encountered during transform,
### the resulting one-hot encoded columns for this feature will be all zeros. 
### In the inverse transform, an unknown category will be denoted as None.

ohe = OneHotEncoder(handle_unknown='ignore')


ohe_df = pd.DataFrame(ohe.fit_transform(df[['class']]).toarray())


final_df = df.join(ohe_df)


final_df.drop('class', axis=1, inplace=True)
final_df
```

# Q3: Apply Frequency Encoding. 

### I found amazing DF to apply Frequency and Binary Encoding

```python
dF = pd.read_excel("https://raw.githubusercontent.com/atulpatelDS/Data_Files/master/Feature_Engineering/feature_encoding.xlsx")
dF.head()
```

```python
dF['CITY'].value_counts()
```

```python
feq = (dF.groupby("CITY").size()) / len(dF)
feq1 = dF.groupby("CITY").size()
dF['Frequency for the city int'] = dF.CITY.map(feq1)
dF['Frequency for the city float'] = dF.CITY.map(feq)
dF
```

# Q4: Apply Binary Encoding. 

```python
cebe = category_encoders.BinaryEncoder(cols=['CUSTOMER_TYPE'])
df_New = cebe.fit_transform(dF['CUSTOMER_TYPE'])
final_df = dF.join(df_New)
final_df
```
